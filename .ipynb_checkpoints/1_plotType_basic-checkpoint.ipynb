```python
import matplotlib.pyplot as plt
import numpy as np
```

<h1>1. plot(x, y)  :  matplotlib.axes.Axes.plot</h1>
<h3>Axes.plot(*args, scalex=True, scaley=True, data=None, **kwargs)</h3><br>
Plot y versus x as lines and/or markers.

```python
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, axs = plt.subplots(1, 2)

axs[0].plot(x, y)  # x와 y를 plotting
axs[1].plot(y)  # x는 default (정수)로 plotting
```

```python
# plot multiple data
x1 = np.linspace(0, 10, 100)
y1 = 4 + 2 * np.sin(2 * x1)
x2 = np.linspace(1, 11, 100)
y2 = 4 + 2 * np.sin(2 * (x2 - 1))

fig, ax = plt.subplots()

ax.plot(x1, y1, x2, y2)
```

```python
# make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

<h2>scatter(x, y)</h2>

```python
import matplotlib.pyplot as plt
import numpy as np

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
